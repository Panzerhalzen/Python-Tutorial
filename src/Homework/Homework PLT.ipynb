```python
import matplotlib.pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] #Года
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3] #ВВП
```

```python
#Обозначим лимиты осей xy
axis_limits = [1940, 2030, 0, 20000]
plt.axis(axis_limits)

#Созадим линейную диаграмму: годы по оси х, ВВП по оси у
# plt.plot(years, gdp, 'ro')
plt.plot(years, gdp, ms=10, color='green', marker='x', linestyle='solid', linewidth=0.5)

#Добивим название диаграммы
plt.title('Номинальный ВВП')

#Добавим посдпись к оси у
plt.ylabel('Млрд $')

#Добивим подпись к оси x
plt.xlabel('Года')

#Добавим сетку
plt.grid(True)

#Отобразить график
plt.show()

#Сохраняем график
plt.savefig('Media/homework_plt.png')
```

```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

#Сгруппировать оценки по десятками, но разместить оценки 90 и выше вместе
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()], list(histogram.values()), 10, color='black', edgecolor='green') #Ширина каждого стобца 10
plt.axis([-5, 105, 0, 5])
plt.xticks([11 * i for i in range(11)]) #Градация шкалы х
plt.xlabel('Дециль')
plt.ylabel('Число студентов')
plt.title('Распределение оценок за экзамен')
plt.show()

plt.savefig('Media/homework_exam_grades.png')
```

```python
coffee = [1, 5, 3, 1.5, 2, 4] #Кружки кофе
minutes = [60, 150, 110, 83, 96, 123] #Минуты без сонливости
labels = ['one', 'five', 'three', 'half', 'two', 'four']

plt.scatter(coffee, minutes, color='black', edgecolor='red')

#Назначим метку для каждой точки
for label, coffee_count, minute_count in zip(labels, coffee, minutes):
    plt.annotate(label, 
                xy=(coffee_count, minute_count), #Задать метку
                xytext=(0, -12.5),                 #Немног осместить метку
                textcoords='offset points')
plt.title('Кол-во кружек кофе против минут без сна')
plt.xlabel('Кол-во кружек кофе')
plt.ylabel('Минут без сна')
plt.show()
```

```python
import math
import random

def squares(n: int) -> int:
    return n ** 2

num_sq = [squares(random.randint(0, 99)) for i in range(1000)]

fig, ax = plt.subplots()
ax.scatter([i for i in range(1000)], num_sq, s=10, c='red')
plt.style.use('seaborn-dark')
plt.grid(True)

plt.show()
```

```python
#plt.subplot()
#Позволяет размещать несколько графиков на разных полях.
#plt.subplot(nrows, ncols, index)
#nrows: int -> количество строк
#ncols: int -> количество столбцов
#index: int -> местоположение элемента

#или

#pls.subplot(pos)
#pos: int -> позиция в виде трехзначного числа, содержащего информацию о количестве строк, столбцов и индексе

x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]

#настройка размеров подложки
plt.figure(figsize=(12, 7))

#вывод графиков
plt.subplot(2, 2, 1)
plt.plot(x, y1, '-')
plt.subplot(2, 2, 2)
plt.plot(x, y2, '-')
plt.subplot(2, 2, 3)
plt.plot(x, y3, '-')
plt.subplot(2, 2, 4)
plt.plot(x, y4, '-')
plt.show()
```

```python
#Второй вариант использования subplot()
plt.figure(figsize=(12, 7))

plt.subplot(221)
plt.plot(x, y1, '-')
plt.subplot(222)
plt.plot(x, y2, '-')
plt.subplot(223)
plt.plot(x, y3, '-')
plt.subplot(224)
plt.plot(x, y4, '-')
plt.show()
```

```python
#Работа с функцией subplots()

#Неудобство использования subplot() - необходимо каждый раз указывать количество строк и столбцов сетки
#Функция subplots() возвращает два объекта:
# - Figure - подложка, на которой будут размещены поля
# - Axes - объект или массив объектов, через которые можно получить полный доступ к настрйоке внешнего вида
#                                                                                    отображаемых элементов

fig, ax = plt.subplots(2, 2, figsize=(12, 7))

#Третьим аргументов указывает тип отображаемой линии
ax[0, 0].plot(x, y1, '-')
ax[0, 1].plot(x, y2, '--')
ax[1, 0].plot(x, y3, '-.')
ax[1, 1].plot(x, y4, ':')
plt.show()
```

```python
fig, ax = plt.subplots()
plt.style.use('seaborn')

#Если не указывать размер подложки и сетку, то все графики будут отображатсья на одном поле
ax.plot(x, y1, '-')
ax.plot(x, y2, '--')
ax.plot(x, y3, '-.')
ax.plot(x, y4, ':')
plt.show()
```
